```python
# converting the data into dependent and independent variables
x = data1.drop(['age','address','famsize','Pstatus','Medu','Fedu','reason','guardian','famsup','paid','nursery','higher','internet',
              'romantic','famrel','schoolsup','passed'],axis=1)
y = data1['passed']
```

# One hot Encoding

```python
x = pd.get_dummies(x, columns=['school','activities','sex'])

y = pd.get_dummies(y, columns=['passed'])
```

```python
x.head()
```

```python
y.head()
```

```python
x['Pjob']=x['Mjob']+['Fjob']
x['Freetimes']=x['freetime']+x['goout']
```

```python
x = x.drop(['freetime','Mjob','Fjob'],axis=1)
```

```python
x.info()
```

```python
x.head()
```

```python
x=x.drop(['Pjob'],axis=1)
```

# Splitting the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

# Building Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import f1_score,confusion_matrix
print('f1_score:',f1_score(y_test,y_pred,average='weighted'))
print('confusion_matrix:',confusion_matrix(y_test.values.argmax(axis=1),y_pred.argmax(axis=1)))
```

f1_score is 61%

# Verifying the Feature Importance

```python
pd.Series(rf.feature_importances_,index=x.columns).sort_values(ascending = True)*100
```

```python
rf.fit(x_train,y_train)
```

```python
rfit= RandomForestClassifier(n_estimators=500,max_depth=10,random_state=42,criterion="entropy")
rfit.fit(x_train,y_train)
```

```python
y_pred=rfit.predict(x_test)
f1_score(y_test,y_pred,average='weighted')
```
